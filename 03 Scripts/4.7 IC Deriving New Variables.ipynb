## 2. Step 1a. Create “price_label” column.

```python
# Create new "price_label" column, define a function:
def price_label(row):
    
    if row['prices'] <=5:
        return 'Low range product'
    elif (row['prices'] >5) and (row['prices'] <=15):
        return 'Medium range product'
    elif row['prices'] >15:
        return 'High range product'
    else: return np.nan
```

```python
# Apply the function
df = df.copy()
df['price_range'] = df.apply(price_label, axis=1)
```

## 3. Step 1b. Create "busiest_day" column.

```python
# Create a new column “busiest day,” that will contain one of three different values: “Busiest day,” “Least busy,” and “Regularly busy".
result = []

for value in ords_prods_merged["orders_day_of_week"]:
  if value == 0:
    result.append("Busiest day")
  elif value == 4:
    result.append("Least busy")
  else:
    result.append("Regularly busy")
```

```python
# Create a new column within ords_prods_merge dataframe and set it equal to result
ords_prods_merged['Busiest_day']=result
```

```python
ords_prods_merged['Busiest_day'].value_counts(dropna=False)
```

## 4. Create “Busiest days” columns = two busiest days of the week, and two slowest days colum.

```python
# Create column "Busiest days" from "orders_day_of_week", to identify two busiest days (0, 1), slowest days (3, 4), or regularly busy (all other days).
result_2 = []

for value in ords_prods_merged["orders_day_of_week"]:
  if value == 0 or value == 1:
    result_2.append("Busiest day")
  elif value == 3 or value == 4:
    result_2.append("Least busy")
  else:
    result_2.append("Regularly busy")
```

## 6. Step 4. Create a new column “busiest_period_of_day”, containings the busiest hours of the day/periods of time labeled “Most orders,” “Average orders,” and “Fewest orders.” 

```python
# Find out on which hour most orders take place via printing the frequency of the “order_hour_of_day” column
ords_prods_merged['order_hour_of_day'].value_counts(dropna=False)
```

```python
# Create column "busiest_period_of_day" from "order_hour_of_day", to identify periods of 
# "Most orders" = 10, 11, 14, 15, 13, 12, 16, 9; 
# "Average orders" = 17, 8, 18, 19, 20, 7, 21, 22;
# and "Fewest orders" = 23, 6, 0, 1, 5, 2, 4, 3.
result_3 = []

for value in ords_prods_merged["order_hour_of_day"]:
  if value in [10, 11, 14, 15, 13, 12, 16, 9]:
    result_3.append("Most orders")
  elif value in [23, 6, 0, 1, 5, 2, 4, 3]:
    result_3.append("Fewest orders")
  else:
    result_3.append("Average orders")
```

```python
result_3
```

```python
# Create new column "busiest_period_of_day" in ords_prods_merge.
ords_prods_merged['busiest_period_of_day'] = result_3
```

```python
# Check output of ords_prods_merged with new "busiest_period_of_day" column.
ords_prods_merged.head()
```
