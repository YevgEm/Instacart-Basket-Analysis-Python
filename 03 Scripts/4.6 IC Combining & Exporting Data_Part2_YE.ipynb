## 4. Step 5. Determine a suitable way to combine the orders_products_combined dataframe with your products data set.

```python
# Import "products_checked.csv"
df_prods=pd.read_csv(os.path.join(path, '02 Data', 'Prepared Data', 'products_checked.csv'), index_col = False)
```

```python
# Check the columns header of df_ords_prods_combined and df_prods
df_ords_prods_combined.head()
```

#### There's a key column 'product_id' observed among these two dataframes, indicating a merge is possible.
#### I decide to choose for combining of both dataframes merge with how=inner argument.

```python
# Drop the column "_merge" and "Unnamed: 0" from df_ords_prods_combined
df_ords_prods_combined=df_ords_prods_combined.drop(['_merge', 'Unnamed: 0'], axis=1)
```

```python
# Check the output
df_ords_prods_combined.shape
```

```python
# Drop the unnecessary column "Unnamed: 0" from df_prods
df_prods = df_prods.drop(['Unnamed: 0'], axis=1)
```

```python
# Check the output
df_prods.shape
```

```python
# Test merge without overwriting
pd.merge(df_ords_prods_combined, df_prods, on = 'product_id', indicator = True)
```

```python
# Merge the updated dataframes
df_ords_prods_merged = df_ords_prods_combined.merge(df_prods, on = 'product_id', indicator = True)
```

```python
# Check the output
df_ords_prods_merged.head()
```

```python
# Check the output
df_ords_prods_merged.shape
```

## 5. Step 6. Confirm the results of the merge using the merge flag.

```python
# Check the output using merge flag
df_ords_prods_merged['_merge'].value_counts()
```

## 6. Step 7. Export the newly created dataframe as ords_prods_merge in a suitable format (taking into consideration the size).

```python
# Export the merged file in pickle format as “orders_products_merged.pkl”.
df_ords_prods_merged.to_pickle(os.path.join(path, '02 Data','Prepared Data', 'orders_products_merged.pkl'))
```
